```r
library(ggplot2)
```

```r
dirpath <- "DrugResponseGenes"

responsed <- read.csv(file.path(dirpath, "functional_annotation.txt"), sep="\t")
```

```r
filtered <- responsed[which(responsed$PValue < 0.01), ]
filtered$Term <- factor(filtered$Term, levels = filtered$Term[order(filtered$PValue, decreasing = TRUE)])

p <- ggplot(filtered, aes(Term, -log(PValue))) + 
    geom_col(fill = "#619CFF") + 
    coord_flip()
print(p)
ggsave(filename = file.path(dirpath, "functional_annotation.png"))
```
